```python
# Import necessary libraries
from sqlalchemy import create_engine, inspect

# Connect to the database
# Replace 'database_name' with the actual name of your database
engine = create_engine('sqlite:///my_data1.db')

# Create an inspector object
inspector = inspect(engine)

# Get the list of table names in the database
table_names = inspector.get_table_names()

# Choose a specific table from the list
# Replace 'table_name' with the actual name of the table you want to retrieve column names from
table_name = 'SPACEXTBL'

# Get the column names of the chosen table
columns = inspector.get_columns(table_name)

# Extract column names from the columns object
column_names = [column['name'] for column in columns]

# Print the column names
print("Column Names:")
print(column_names)
```

```python
%sql SELECT PAYLOAD \
FROM SPACEXTBL \
WHERE LANDING_OUTCOME = 'Success (drone ship)' \
AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000;
```

```python
%sql SELECT MISSION_OUTCOME, COUNT(*) as total_number \
FROM SPACEXTBL \
GROUP BY MISSION_OUTCOME;
```

```python
%sql SELECT BOOSTER_VERSION \
FROM SPACEXTBL \
WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL);
```

```python
%sql SELECT substr(Date,4,2) as month, DATE,BOOSTER_VERSION, LAUNCH_SITE, [Landing_Outcome] \
FROM SPACEXTBL \
where [Landing_Outcome] = 'Failure (drone ship)' and substr(Date,7,4)='2015';
```

```python
%sql SELECT [Landing_Outcome], count(*) as count_outcomes \
FROM SPACEXTBL \
WHERE DATE between '04-06-2010' and '20-03-2017' group by [Landing_Outcome] order by count_outcomes DESC;
```
